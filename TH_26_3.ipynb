```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/bank-additional-full.xlsx')
```

```python
df.shape
```

```python
df['conversion'] = df['y'].apply(lambda x:1 if x == 'yes' else 0)
```

```python
df
```

```python
df.conversion.sum() #4640/41188
```

```python
print('Conversion rate: %0.2f' % (df.conversion.sum() / df.shape[0] * 100))
#Conversion rate = tỷ lệ chuyển đổi
```

```python
pd.DataFrame(df.groupby(by = 'campaign')['conversion'].sum())
#group by: tập hợp - min, max, sum, count
#Đếm theo campaign
#Lấy các giá trị bằng 1
```

```python
pd.DataFrame(df.groupby(by = 'campaign')['conversion'].count())
#ứng với từng campaign có bao nhiêu khách hàng
#đếm số
```

```python
#tính conversion rate theo từng campaign
conversion_by_campaign = pd.DataFrame(df.groupby('campaign')['conversion'].count()/
                                     df.groupby('campaign')['conversion'].sum())
```

```python
conversion_by_campaign
```
